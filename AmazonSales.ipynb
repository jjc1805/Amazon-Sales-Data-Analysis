#Amazon Sales Analysis
This project is to analyze the Amazon Sales data and obtain meaninful relationships within the data provided. The main target will be to find the Sales trend -> month-wise, year-wise and yearly-month wise.

#Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#Importing the dataset

```python
df = pd.read_csv('/content/Amazon Sales data.csv')

# Print the DataFrame
df.head()
```

```python
# Displaying the information in the dataset
df.info()
```

#Sales Trend

1. Month wise

```python
# Month Wise
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Month'] = df['Order Date'].dt.month
df.groupby('Month')['Total Profit'].sum().plot(kind='bar', figsize=(12,7))
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.title("Total Profit by Month")
plt.show()
```

2. Year wise

```python
# Year Wise
df['Year'] = df['Order Date'].dt.year
df.groupby('Year')['Total Profit'].sum().plot(kind='bar', figsize=(12,7))
plt.xlabel("Year")
plt.ylabel("Total Profit")
plt.title("Total Profit by Year")
plt.show()
```

3. Yearly Month wise

```python
# Yearly Month wise
df.groupby(['Year', 'Month'])['Total Profit'].sum().unstack().plot(kind='bar', figsize=(14,8))
plt.xlabel("Yearly-Months")
plt.ylabel("Total Profit")
plt.title("Total Profit by Month and Year")
plt.xticks(rotation=0)
plt.show()
```

#Total Profit by Order Date

```python
df.groupby('Order Date')['Total Profit'].sum().plot(kind='hist', figsize=(12,7))
plt.xlabel("Order Date")
plt.ylabel("Total Profit")
plt.title("Total Profit by Order Date")
plt.show()
```
